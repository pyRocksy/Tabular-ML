## Data Scientist Associate Practical Exam Submission

**Project Background**
- GoalZone is a fitness club chain in Canada.
- GoalZone offers a range of fitness classes in two capacities - 25 and 15.
- Some classes are always fully booked. Fully booked classes often have a low attendance rate.

**Project Goal**
- GoalZone wants to increase the number of spaces available for classes.
- They want to do this by predicting whether the member will attend the class or not.
- If they can predict a member will not attend the class, they can make another space available.

**Classification ML Algorithms Used**
- Logistic Regression
- XGBoost with SMOTE

## Dataset
The dataset contains the attendance information for the class scheduled this year so far. The data you will use for this analysis can be accessed here: [`fitness_class`]([https://s3.amazonaws.com/talent-assets.datacamp.com/fitness_class_2212.csv])

## Scaling and Normalizing `months_as_member`, `weight`, `days_before` columns


**Scaling `months_as_member` using StandardScaler, MinMaxScaler and Numpy Log Transformation**

```python
import matplotlib.pyplot as plt
import numpy as np 
from sklearn.preprocessing import StandardScaler,MinMaxScaler

# Log Transform
df["Log_months_as_member"] = np.log(df["months_as_member"] + 1)

# MinMax Sclaer
scaler = MinMaxScaler()
df["MinMax_Scaled_months_as_member"] = scaler.fit_transform(df[["months_as_member"]])

# Standard Scaler
scaler = StandardScaler()
df["SC_Scaled_months_as_member"] = scaler.fit_transform(df[["months_as_member"]])

# Create the histogram subplot
fig, ax = plt.subplots(1, 2, figsize=(12,6))
ax[0].hist(df["months_as_member"], bins=20)
ax[0].set_title("Before Scaling")
ax[1].hist(df["Log_months_as_member"], bins=20)
ax[1].set_title("After Scaling")

plt.suptitle("Fig 20a: Log Scaling 'months_as_member' column")
# Show the plot
plt.show()

# Create the histogram subplot
fig, ax = plt.subplots(1, 2, figsize=(12,6))
ax[0].hist(df["months_as_member"], bins=20)
ax[0].set_title("Before Scaling")
ax[1].hist(df["SC_Scaled_months_as_member"], bins=20)
ax[1].set_title("After Scaling")

plt.suptitle("Fig 20b: Stanardrd Scaling 'months_as_member' column")
plt.show()

# Create the histogram subplot
fig, ax = plt.subplots(1, 2, figsize=(12,6))
ax[0].hist(df["months_as_member"], bins=20)
ax[0].set_title("Before Scaling")
ax[1].hist(df["MinMax_Scaled_months_as_member"], bins=20)
ax[1].set_title("After Scaling")

plt.suptitle("Fig 20c: MinMax Scaling 'months_as_member' column")
plt.show()
```

```python
# Load required modules
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split,cross_val_score
from sklearn.metrics import confusion_matrix, precision_score, recall_score, auc, roc_curve, roc_auc_score, accuracy_score, classification_report
```

**Split Data, Fit and Predict Logistic Regression Model**

Since class 0 represnts 69.73% of "attended" column and class 1 represents 30.27%, I am setting the following hyperparameters to account for class imbalance.

## GridSearch XGBoost

```python
# Import necessary modules
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
from xgboost import XGBClassifier
from imblearn.over_sampling import SMOTE

# Define X and y
X = df[best_features]
y = df['attended']

# Split the data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.2, random_state=123)

# Perform SMOTE oversampling
sm = SMOTE(random_state=123)
X_train_sm, y_train_sm = sm.fit_resample(X_train, y_train)

# Create a pipeline for XGBoost Classifier
pipe_xgb = Pipeline([('xgb', XGBClassifier(random_state=13))])

# Define hyperparameter grid for XGBoost Classifier
param_grid_xgb = {'xgb__max_depth': [3, 5, 7, 10],
                  'xgb__learning_rate': [0.1, 0.01, 0.001],
                  "xgb__n_estimators": [20,50,150, 200]}

# Create a GridSearchCV object
grid_xgb = GridSearchCV(pipe_xgb, param_grid_xgb, cv=10)

# Fit the GridSearchCV object to the data
grid_xgb.fit(X_train, y_train)

# Print the best hyperparameters and mean cross-validation score for XGBoost Classifier
print("XGBoost Classifier best parameters: ", grid_xgb.best_params_)
print("XGBoost Classifier mean CV score: ", grid_xgb.best_score_)
```

### Model 2: `XGBoost Classifier` with SMOTE
- `Due to class imbalance, I am using SMOTE to oversample the minoity group.`
- `After grid searching, the following hyperparameters achieved best results.` 
	- `learning_rate=0.01, max_depth= 3, n_estimators=150`

```python
# load required libraries
from xgboost import XGBClassifier
from imblearn.over_sampling import SMOTE
from sklearn.metrics import accuracy_score, confusion_matrix, recall_score, precision_score
import seaborn as sns
import matplotlib.pyplot as plt

# Define X and y
# X = df.drop(["attended", "booking_id"], axis=1)
X = df[best_features]
y = df['attended']

# Split the data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.2, random_state=123)

# Perform SMOTE oversampling
sm = SMOTE(random_state=123)
X_train_sm, y_train_sm = sm.fit_resample(X_train, y_train)

# Create XGBoost model
model_xgb = XGBClassifier(random_state=13, learning_rate=0.01, max_depth= 3, n_estimators=150)

# Fit the model to the upsampled training data
model_xgb.fit(X_train_sm, y_train_sm)

# Predict on the test data
y_pred_xgb = model_xgb.predict(X_test)

# Evaluating Model Performance
XGB_CV_Scores = cross_val_score(model_xgb, X_train_sm, y_train_sm, cv=5)
XGB_Accuracy = accuracy_score(y_test, y_pred_xgb)
XGB_Recall_test_set = recall_score(y_test, y_pred_xgb) 
XGB_ROC_AUC_Score = roc_auc_score(y_test, y_pred_xgb)
XGB_Classification_Report = classification_report(y_test, y_pred_xgb)
XGB_Confusion_Matrix = confusion_matrix(y_test, y_pred_xgb)
```
